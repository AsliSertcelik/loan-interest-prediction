```python
clc;
clear;

veri yükleme
data = readtable('train_loan.csv');

head(data)

eksik veri temizleme

% Months_Since_Delinquency → 0 yap
data.Months_Since_Deliquency(ismissing(data.Months_Since_Deliquency)) = 0;

% Annual_Income → median
med = median(data.Annual_Income,'omitnan');
data.Annual_Income(ismissing(data.Annual_Income)) = med;

log dönüşümü
data.Annual_Income = log(data.Annual_Income + 1);

one hot encoding

categorical_vars = ["Home_Owner","Income_Verified","Purpose_Of_Loan","Gender","Length_Employed"];

for i = 1:length(categorical_vars)

    col = categorical_vars(i);

    if any(strcmp(data.Properties.VariableNames, col))

        data.(col) = categorical(data.(col));

        cats = categories(data.(col));

        dummy = dummyvar(data.(col));

        dummy_table = array2table(dummy);

        dummy_table.Properties.VariableNames = strcat(col, "_", cats);

        data = [data dummy_table];

        data.(col) = [];

    end
end

model hazırlık

X = data;
Y = data.Interest_Rate;

X.Interest_Rate = [];
X_numeric = varfun(@isnumeric, X, 'OutputFormat', 'uniform');
X{:, X_numeric} = normalize(X{:, X_numeric});

Train-test split

cv = cvpartition(height(data),'HoldOut',0.2);

X_train = X(training(cv),:);
X_test = X(test(cv),:);

Y_train = Y(training(cv));
Y_test = Y(test(cv));

Logistic regression

model = fitcecoc(X_train, Y_train);

tahmin

Y_pred = predict(model, X_test);

accuracy

accuracy = sum(Y_pred == Y_test) / length(Y_test);
disp(accuracy)

confusion matrix

confusionchart(Y_test, Y_pred)

```
